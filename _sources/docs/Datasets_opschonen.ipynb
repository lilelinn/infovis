## Dataset schietincidenten

```python
import pandas as pd
gv = pd.read_csv('stage3.csv')
gv.head()
```

Om deze dataset schoon te maken is het van belang om te filteren per jaar. De andere dataset beval namelijk informatie tot en met 2017, waar deze dataset door gaat tot 2018. Om dit op te lossen maken we eerst een aparte dataframe voor de relevante jaren. Door deze samen te voegen ontstaat de nieuwe, schoongemaakte dataset. 

Daarnaast staat 'District of Columbia' geregistreerd als area. Dit is daarentegen geen staat, dus deze moet ook verwijderd worden.

Om dit te kunnen doen moet eerst de 'date' kolom in de dataset geconverteerd worden naar een datetime format, zodat het mogelijk is om te filteren op deze velden.

```python
# converteer de 'date' kolom naar datetime
gv['date'] = pd.to_datetime(gv['date'])

# verwijder district of columbia
gv = gv[gv.state != 'District of Columbia']
```

```python
# aparte dataframes voor de jaren 2013-2017 uit de dataset
gv13 = gv[gv['date'].dt.year == 2013]
gv14 = gv[gv['date'].dt.year == 2014]
gv15 = gv[gv['date'].dt.year == 2015]
gv16 = gv[gv['date'].dt.year == 2016]
gv17 = gv[gv['date'].dt.year == 2017]
```

Nu we de dataframes hebben voor elk individueel jaar, voegen we deze samen. Dit doen we via concat, zodat alle dataframes onder elkaar staan op chronologische volgorde.

```python
# schone dataset is een combinatie van de jaren 2013 - 2017
gv_cleaned = pd.concat([gv13, gv14, gv15, gv16, gv17], axis=0)
```

```python
gv_cleaned.head()
```

```python
# exporteren
gv_cleaned.to_csv('gvcleaned.csv')
```

## Dataset BBP

```python
gdp = pd.read_csv("bea-gdp-by-state.csv")
gdp.head()
```

Voor het opschonen van deze dataset was het belangrijk om irrelevante kolommen en rijen te verwijderen. Zo was de kolom 'Fips' irrelevant voor ons onderzoek. Daarnaast bevatte de dataset informatie over bepaalde regio's in de Verenigde Staten die niet kwalificeerden als staat of voorkomen in onze andere dataset. Deze hebben we via 'drop' verwijderd. 

```python
# drop irrelevante kolommen en rijen
gdp = gdp.drop(columns=['Fips'])
gdp = gdp.drop([0, 9, 52, 53, 54, 55, 56, 57, 58, 59])
```

```python
gdp.head()
```

## Dataset geweren

```python
gc = pd.read_csv('us_gun_production1.csv')
gc.head()
```

Bij het schoonamaken van deze dataset kijken we vooral naar het toevoegen van een kolom: het totaal aantal geweren in circulatie in de VS per jaar. Om dit te bereiken moet een berekening gemaakt worden: het totaal geproduceerd - het totaal geÃ«xporteerd + het totaal geimporteerd. Voor deze berekeningen moeten de punten en komma's die momenteel in die velden staan om duizentallen aan te geven weggehaald worden. Pas dan kan de berekening plaatsvinden.

Verder wordt de rij met informatie over de jaren 1899 - 1945 verwijderd. Dit is een samenvoeging van 46 jaar, en is daardoor een enorme uitschieter ten opzichte van de overige jaren. Verder maken we de kolom met jaren de index van de dataset.

```python
# set index en verwijder 1899 - 1945
gc = gc.set_index('Year')
gc = gc.drop(['1899-1945'])

# kolommen bereid maken voor berekeningen d.m.v. replace
gc['Total Imported'] = gc['Total Imported'].replace(',','',regex=True)
gc['Total Imported'] = gc['Total Imported'].replace('\.','',regex=True)

gc['Total Manufactured'] = gc['Total Manufactured'].replace(',','',regex=True)
gc['Total Exported'] = gc['Total Exported'].replace('\.','',regex=True)
gc['Total Exported'] = gc['Total Exported'].replace(',','',regex=True)

# berekeningen totaal aantal geweren in de VS per jaar
gc['Total Circulated'] = gc['Total Manufactured'].astype(int) - gc['Total Exported'].astype(int) + gc['Total Imported'].astype(int)
```

```python
gc.head()
```

```python
# exporteren
gc.to_csv('gc_cleaned.csv')
```

## Dataset uitgaven BBP

```python
imfdf = pd.read_csv('imfdm.csv')
imfdf.head()
```

Om deze dataset schoon te maken moet er vrij weinig gebeuren. Alles wat we aanpassen is de index (deze zetten we eerst naar Country, maar zal later veranderen) en we draaien de dataframe om, zodat de jaren als rijen staan. Dit omdat de dataframe dan dezelfde structuur heeft als de eerdere datasets. 

Na het omdraaien zal de index veranderen van de landen naar de jaren. Zo heeft deze dataframe (op lengte na) dezelfde index als de dataset hiervoor.

```python
# zet index en draai dataframe om
imfdf = imfdf.set_index('Country')
imfdf = imfdf.T
```
